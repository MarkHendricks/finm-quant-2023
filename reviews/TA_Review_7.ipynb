```python
import pandas as pd

df = pd.DataFrame(
    {
        'A': [i for i in range(50)],
        'B': [i for i in range(50, 100)],
        'C': [i for i in range(100, 150)]
    }
)
```

```python
df.pct_change().head(10)
```

```python
df.diff().head(10)
```

```python
df.describe()
```

```python
# Filtering
df_filt = df[df['A'] < 25]
df_filt
```

```python
# Or,
df_filt = df.loc[df['A'] > 25, ['A', 'C']]
df_filt
```

```python
import numpy as np 

df_timeindexed = pd.DataFrame(
    data = np.random.randn(1000, 4),
    index=pd.date_range('1/1/2000', periods=1000),
)

# Make index a datetime index
df_timeindexed.index = pd.to_datetime(df_timeindexed.index)

df_timeindexed.resample('M').mean()
```

```python
(df_timeindexed.resample('M').first() - df_timeindexed.resample('M').last())/df_timeindexed.resample('M').first()
```

```python
df_timeindexed.rolling(100).std() * 10
```
